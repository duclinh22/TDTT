Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và chiều cao (kiểu số thực) với hai phương thức tính diện tích bề mặt (diện tích toàn phần) và thể tích của hình trụ

Đầu vào
Đầu vào từ bàn phím gồm một dòng duy nhất chứa bán kính và chiều cao của hình trụ, cách nhau bởi một dấu cách.

Đầu ra
In ra màn hình diện tích bề mặt và thể tích của hình trụ (sử dụng lại những hàm vừa viết, $\Pi=3.14$). (Kết quả làm tròn đến 2 chữ số sau dấu phẩy).

```python
class Cylinder:
    def __init__(self,bankinh,chieucao):
        self.h=chieucao
        self.r=bankinh
    def S(self):
        return 2*3.14*self.r*(self.r+self.h)
    def V(self):
        return 3.14*(self.r**2)*self.h
r,h=map(float,input().split())
hinhtru=Cylinder(r,h)
dientich=hinhtru.S()
thetich=hinhtru.V()
print(f'{dientich:.2f} {thetich:.2f}')
```

Đầu vào
Đầu vào từ bàn phím chứa dữ liệu ngày, tháng, năm dưới dạng dd/mm/yyyy

Đầu ra
In ra màn hình ngày kế tiếp của ngày vừa nhập dưới dạng dd/mm/yyyy
Nếu ngày tháng không tồn tại, in ra INVALID

```python
class tinhngay():
    def __init__(self,ngay,thang,nam):
        self.d=ngay
        self.m=thang
        self.y=nam
    def maxday(self):
        match self.m:
            case 1|3|5|7|8|10|12:
                maxday=31
            case 4|6|9|11:
                maxday=30
            case 2:
                if (self.y%4==0 and self.y%100!=0) or self.y%400==0:
                    maxday=29
                else:
                    maxday=28
            case _:
                maxday=-1
        return maxday
    def hople(self):
        max=self.maxday()
        if  1<=self.d<=max:
            return True
        return False
    def nextday(self):
        max=self.maxday()
        if not self.hople():
            return 'INVALID'
        if self.d<max:
            self.d+=1
        elif self.d==max:
            if self.m<12:
                self.d=1
                self.m+=1
            elif self.m==12:
                self.d=1
                self.m=1
                self.y+=1
        return f'{self.d}/{self.m}/{self.y}'
d,m,y=map(int,input().split('/'))
ngay=tinhngay(d,m,y)
print(ngay.hople())
print(ngay.nextday())

```

- scale(k) phóng to hình chữ nhật theo hệ số k (nhân cả w và h), không trả về gì.

Input
w h k (số thực dương)

Output

```python
class Rectangle():
    def __init__(self,w,h):
        self.w=w
        self.h=h
    def area(self):
        return self.w * self.h
    def perimeter(self):
        return 2 * (self.w + self.h)
    def scale(self,k):
        self.w *= k
        self.h *= k
w,h,k=map(float,input().split())
hcn=Rectangle(w,h)
hcn.scale(k)
print(hcn.area())
print(hcn.perimeter())
```

- avg() trả về điểm trung bình.
- rank() trả về "Excellent" nếu avg ≥ 8, "Good" nếu 6.5-<8, "Average" nếu 5-<6.5, ngược lại "Poor".

Input
Dòng 1: name
Dòng 2: n
n dòng tiếp: subject score (score là số thực 0-10)

Output

```python
class Student():
    def __init__(self,name):
        self.name=[]
        self.ten=name
    def add_score(self,subject,score):
        self.name.append({'mon':subject,'diem':score})
    def avg(self):
        count=0
        diem=0
        for i in self.name:
            count+=1
            diem+=i['diem']
        if count==0:
            return 0
        return diem/count
    def rank(self):
        if Student.avg(self)>=8:
            return 'Excellent'
        elif Student.avg(self)>=6.5:
            return 'Good'
        elif Student.avg(self)>=5:
            return 'Average'
        else:
            return 'Poor'
name=input('Nhap ten hoc sinh')
n=int(input('Nhap so mon'))
stud=Student()
for i in range(n):
    mon,diem=input('Nhap mon va diem').split()
    diem=int(diem)
    stud.add_score(mon,diem)
print(name)
print(f'{stud.avg():.2f}')
print(stud.rank())
```

**W8A11**

Bạn được cho bốn điểm A, B, C và D trong hệ tọa độ Đề-các 3 chiều.
Nhiệm vụ của bạn là tính và in ra góc giữa hai mặt phẳng tạo bởi các điểm:
- mặt phẳng (A, B, C)
- mặt phẳng (B, C, D)
Góc cần in ra được ký hiệu là PHI, và phải dùng độ (degrees), không dùng radian.

Dạng Input:
Một dòng input chứa 3 số thực, ngăn cách bằng dấu cách, biểu diễn: tọa độ X tọa độ Y tọa độ Z của một điểm.

Dạng Output:
In ra góc giữa hai mặt phẳng, chính xác tới hai chữ số thập phân.

```python
import math

class Vector:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y, self.z - other.z)

    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z

    def cross(self, other):
        return Vector(
            self.y * other.z - self.z * other.y,
            self.z * other.x - self.x * other.z,
            self.x * other.y - self.y * other.x
        )

    def absolute(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

if __name__ == '__main__':
    points = []
    for _ in range(4):
        coords = list(map(float, input().split()))
        points.append(coords)

    A = Vector(*points[0])
    B = Vector(*points[1])
    C = Vector(*points[2])
    D = Vector(*points[3])

    AB = B - A
    BC = C - B
    CD = D - C

    n1 = AB.cross(BC)
    n2 = BC.cross(CD)

    cos_phi = n1.dot(n2) / (n1.absolute() * n2.absolute())
    cos_phi = max(min(cos_phi, 1.0), -1.0)

    phi_rad = math.acos(cos_phi)
    phi_deg = math.degrees(phi_rad)

    print(f"{phi_deg:.2f}")
```

[Complex]
Khai báo kiểu dữ liệu biểu diễn số phức ComplexNumber bao gồm 2 biến realPart và imaginaryPart kiểu double biểu diễn phần thực và phần ảo của một số phức.
Viết hàm khởi tạo ComplexNumber(double initRealPart, double initlmaginaryPart) để khởi tạo các giá trị phần thực và phần ảo của một số phức và hàm ComplexNumber multiply(ComplexNumber other) trả về tích của số phức hiện tại và một số phức khác.

Đầu vào
Đầu vào từ bàn phím gồm hai dòng, mỗi dòng gồm hai số thực cách nhau bởi một dấu cách biểu diễn phần thực và phần ảo của một số phức. Các số trong đầu vào có giá trị tuyệt đối không vượt quá 100.

Đầu ra
In ra màn hình hai số biểu diễn phần thực và phần ảo của tích hai số phức.

```python
class ComplexNumber:
    def __init__(self, initRealPart, initImaginaryPart):
        self.realPart = initRealPart
        self.imaginaryPart = initImaginaryPart

    def multiply(self, other):
        new_real = (self.realPart * other.realPart) - (self.imaginaryPart * other.imaginaryPart)
        new_imag = (self.realPart * other.imaginaryPart) + (self.imaginaryPart * other.realPart)
        return ComplexNumber(new_real, new_imag)

if __name__ == '__main__':
    real1, imag1 = map(float, input().split())
    real2, imag2 = map(float, input().split())

    c1 = ComplexNumber(real1, imag1)
    c2 = ComplexNumber(real2, imag2)

    result = c1.multiply(c2)

    print(f"{result.realPart} {result.imaginaryPart}")
```

In ra màn hình diện tích tam giác (làm tròn đến 2 chữ số thập phân sau dấu phẩy). In ra invalid nếu 3 số nhập vào không phải là 3 cạnh của một tam giác

```python
class Triangle():
    def __init__(self,a1,a2,a3):
        self.a1=int(a1)
        self.a2=int(a2)
        self.a3=int(a3)
    def check(self):
        if self.a1+self.a2>self.a3 and self.a1+self.a3>self.a2 and self.a2+self.a3>self.a1:
            return True
        return False
    def getArea(self):
        if not self.check():
            return -1
        p=(self.a1+self.a2+self.a3)/2
        return (p*(p-self.a1)*(p-self.a2)*(p-self.a3))**0.5
a1,a2,a3=input().split()
tamgiac=Triangle(a1,a2,a3)
print(f'{tamgiac.getArea():.2f}')
```

Yêu cầu
Xây dựng lớp Animal với phương thức speak().
Xây dựng hai lớp kế thừa Dog và Cat, mỗi lớp ghi đè speak() để trả về âm thanh tương ứng:
- Dog -> Woof
- Cat -> Meow

Input
Một dòng chứa tên lớp: Dog hoặc Cat.

Output
In ra kết quả của speak().

```python
class Animal():
    def speak():
        pass
class Dog(Animal):
    def speak():
        print('Woof')
class Cat(Animal):
    def speak():
        print('Meow')
type=input()
if type=='Dog':
    Dog.speak()
else:
    Cat.speak()
```

**W8A15**

Yêu cầu
Xây dựng lớp Employee(name, salary).
Xây dựng lớp Manager(name, salary, bonus) kế thừa từ Employee.
Cả hai lớp đều có phương thức income():
- Employee: trả về salary
- Manager: trả về salary + bonus

Input
Dòng 1: loại đối tượng (Employee hoặc Manager)
Dòng 2: name salary [bonus]

Output
In ra tổng thu nhập của đối tượng.

```python
class Employee():
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def income(self):
        return self.salary
class Manager(Employee):
    def __init__(self,name,salary,bonus):
        super().__init__(name,salary)
        self.bonus=bonus
    def income(self):
        return self.salary+self.bonus
type=input()
if type=='Employee':
    name,salary=input().split()
    salary=int(salary)
    a1=Employee(name,salary)
    print(a1.income())
else:
    name,salary,bonus=input().split()
    salary=int(salary)
    bonus=int(bonus)
    a2=Manager(name,salary,bonus)
    print(a2.income())
```
