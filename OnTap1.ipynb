**Bài 2**

Yêu cầu: Cho số nguyên x, trả về True nếu tồn tại số nguyên n>=1 sao cho x=1+2+...+n, ngược lại trả về False. Khuyến khích dùng vòng lặp đơn giản để cộng dồn tổng.

```python
x=int(input())
def kiemtra(x):
    if x<1:
        return False
    else:
        tong=0
        for i in range(1,x):
            tong+=i
            if tong==x:
                return True
            elif tong>x:
                return False
print(kiemtra(x)) 
```

**Bài 4**

Yêu cầu: Cho các hệ số thực a, b, c, trả về tất cả các nghiệm thực của phương trình bậc hai ax^2+bx+c=0 trong một danh sách được sắp xếp tăng dần.
Quy tắc:
- Nếu a==0 và b!=0: trả về nghiệm duy nhất [-c/b].
- Nếu a==0 và b==0: trả về [] (không có nghiệm hữu hạn duy nhất nếu c==0, hoặc vô nghiệm nếu c!=0).
- Với a!=0, dùng biệt thức d = b*b - 4*a*c:
  - d < 0 -> []
  - d == 0 -> [-b/(2*a)]
  - d > 0 -> hai nghiệm [r1, r2] với r1<=r2.

```python
a,b,c=map(float,input().split())
if a==0 and b!=0:
    print(f'Phuong trinh co nghiem duy nhat = {-c/b}')
elif a==0 and b==0:
    if c!=0:
        print('Phuong trinh vo so nghiem')
    elif c!=0:
        print('Phuong trinh vo nghiem')
elif a!=0:
    delta=b**2-4*a*c
    if delta<0:
        print('Phuong trinh vo nghiem')
    elif delta==0:
        print(f'Phuong trinh co nghiem kep={-b/(2*a)}')
    else:
        print(f'Phuong trinh co hai nghiem phan biet x1={(-b-delta**0.5)/(2*a)} va x2={(-b+delta**0.5)/(2*a)}')
    
```
