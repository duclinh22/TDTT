Input
balance rate

Output
Số dư sau khi cộng lãi, in 2 chữ số thập phân.

```python
class Account():
    def __init__(self,balance):
        self.sodu=balance
    def deposit(self,x):
        self.sodu-=x
    def get_balance(self):
        return self.sodu
class SavingAccount(Account):
    def __init__(self,balance,rate):
        super().__init__(balance)
        self.rate=rate
    def add_interest(self,rate):
        self.sodu+=self.sodu*self.rate
balance,rate=list(map(float,input().split()))
taikhoan=SavingAccount(balance,rate)
taikhoan.add_interest(rate)
print(f'{taikhoan.get_balance():.2f}')
```

[a b]
[c d]
Viết phương thức trace() trả về $a+d$ và determinant() trả về ad - bc.

Input
Một dòng chứa 4 số nguyên a b c d.

Output
In ra trace và determinant trên cùng một dòng.

```python
class Matrix2x2():
    def __init__(self,a,b,c,d):
        self.a=a
        self.b=b
        self.c=c
        self.d=d
    def trace(self):
        return self.a+self.d
    def determinant(self):
        return self.a*self.d-self.b*self.c
a,b,c,d=map(int,input().split())
matran=Matrix2x2(a,b,c,d)
print(matran.trace(),matran.determinant())
    
```

Xây dựng lớp Counter có thuộc tính count ban đầu bằng 0 và các phương thức inc(), dec(), value().
Sau đó thực hiện dãy thao tác trên đối tượng.

Input
Dòng 1: n
Dòng 2: n thao tác, mỗi thao tác là INC hoặc DEC, cách nhau bởi dấu cách.

Output
Giá trị cuối cùng của count.

```python
class Counter():
    def __init__(self,count=0):
        self.count=count
    def inc(self,n):
        self.count+=n
    def dec(self,n):
        self.count-=n
    def value(self):
        print(self.count)
n=int(input())
c=Counter()
for i in range(n):
    lenh,n=input().split()
    n=int(n)
    if lenh=='INC':
        c.inc(n)
    elif lenh=='DEC':
        c.dec(n)
c.value()
```

Yêu cầu
Xây dựng lớp Stack sử dụng danh sách để cài đặt ngăn xếp với các phương thức push(x), pop(), top().
Dữ liệu đảm bảo mọi lệnh pop và top đều hợp lệ.

Input
Dòng 1: n
In dòng tiếp theo là một trong ba lệnh:
PUSH x
POP
TOP

Output
Mỗi lệnh TOP in ra giá trị ở đỉnh ngăn xếp trên một dòng.

```python
class Stack():
    def __init__(self):
        self.lst=[]
    def push(self,x):
        self.lst.append(x)
    def pop(self):
        self.lst.pop()
    def top(self):
        return self.lst[-1]
def main():
    s=Stack()
    n=int(input())
    for _ in range(n):
        lenh=list(input().split())
        if lenh=='PUSH':
            x=int(lenh[1])
            s.push(x)
        if lenh=='POP':
            s.pop()
        if lenh == 'TOP':
            print(s.top())
main()
```
