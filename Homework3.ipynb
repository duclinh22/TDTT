### Bài 16
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n=int(input())
def layhangdonvi(n):
    n=n-(n//10)*10
    return n
mu=0
sochan=0
sole=0
while n>(10**mu):
    mu+=1
for i in range(1,mu+1):
    if layhangdonvi(n)%2==0:
        sochan+=1
    else:
        sole+=1
    n=n//10
print(sochan,sole)

```

### Bài 18
Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không

```python
n=int(input())
mu=0
while n>=(3**(mu)):
    mu+=1
muthat=mu-1
if n%(3**(muthat))==0:
    print('YES')
else:
    print('NO')

```

### Bài 19
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó.
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
n=int(input())
a=1
b=1
while a<=n:
    a+=b
    a,b=b,a
print(b-a)

```
