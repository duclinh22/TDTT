### W6A16
Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
m,n=map(int,input('Nhap so hang va so cot').split())
lst=[]
for i in range(n):
    data=list(map(int,input(f'Nhap du lieu cua hang {i+1} ({n} so)').split()))
    lst.append(data)
for i in lst:
    for j in i:
        print(f'{j:>4}',end='')
    print()
```

### W6A17
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và đường chéo phụ của mảng

```python
n=int(input('Nhap vao so hang va so cot'))
lst=[]
for i in range(n):
    data=list(map(int,input(f'Nhap vao hang {i+1}').split()))
    lst.append(data)
print('Ma tran nhap vao la:',lst)
print('Duong cheo chinh la :')
for i in range(n):
    for j in range(n):
        if i==j:
            a=lst[i]
            b=a[i]
            print(b,end=' ')
print()
print('Duong cheo phu la:')
for i in range(n):
    for j in range(n):
        if i+j==n-1:
            a=lst[i]
            b=a[j]
            print(b,end=' ')
```

### W6A18
Nhập vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột k (hàng k)

```python
m,n,k=map(int,input('Nhap vao so cot,so hang,so k').split())
lst=[]
for i in range(n):
    data=list(map(int,input().split()))
    lst.append(data)
#Tinh tong cac phan tu tren cot k:
print(f'Tong cac phan tu tren cot {k} la:')
tongcot=0
for i in lst:
    for j in range(m):
        if j+1==k:
            tongcot+=i[j]
print(tongcot)
#Tinh tong cac phan tu tren hang k:
tonghang=0
print(f'Tong cac phan tu tren hang {k}:')
for i in lst[k]:
    tonghang+=i
print(tonghang)
```

### W6A20-1
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

```python
lst=list(map(int,input().split()))
k=int(input())
def check(lst,k):
    if k not in lst:
        return -1
    giatri=lst.index(k)
    return giatri
print(check(lst,k))
```

### W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
data=input().split()
dic={}
for index,word in enumerate(data):
    if word not in dic:
        dic[word]=[]
    dic[word].append(index)    

dic2={}
for word,indices in dic.items():
    dic2[word]=tuple(indices)
print(dic2)
```

### W6A26
Tính tông các phân tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối

```python
data=list(map(int,input().split()))
tong=0
for i in range(1,len(data)-1):
    if data[i-1]<data[i] and data[i]>data[i+1]:
        tong+=data[i]
print(tong)
```

### W6A27
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). nếu không có in ra -1

```python
data=list(map(int,input().split()))
x=int(input())
lst=[]
for i in data:
    if i<x:
        lst.append(i)
if not lst:
    print('-1')
else:
    print(lst)
```
