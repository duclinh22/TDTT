**Bài 3:** Tính tam giác: Nhập vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
a , b , c = map(float,input().split())
pi=float(3.14)
if a+b>c and a+c>b and b+c>a:
    dientich=(p*(p-a)*(p-b)*(p-c))**(1/2)
    chuvi=a+b+c
    p=chuvi/2
    print(f'Chu vi cua tam giac la {chuvi} , dien tich cua tam giac la {dientich}')
    if a==b==c:
        print('tam giac deu')
    elif a==b or a==c or b==c:
        if a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
            print('tam giac vuong can')
        else:
            print('tam giac can')
    elif a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
        print('Tam giac vuong')
    else:
        print('Tam giac thuong')
else:
    print('Khong phai ba canh cua tam giac')
```

**Bài 5:** Viết chương trình giải phương trình bậc 2 $(ax^{2}+bx+c=0)$

```python
a, b, c = map(float, input().split())

if a == 0:
    if b == 0:
        if c == 0:
            print('Phuong trinh co vo so nghiem')
        else:
            print('Phuong trinh vo nghiem')
    else:
        print(f'Phuong trinh co 1 nghiem: x = {-c/b}')
else:
    delta = b**2 - 4*a*c
    if delta > 0:
        x1 = (-b + delta**0.5) / (2*a)
        x2 = (-b - delta**0.5) / (2*a)
        print(f'Phuong trinh co 2 nghiem phan biet: x1 = {x1}, x2 = {x2}')
    elif delta == 0:
        x = -b / (2*a)
        print(f'Phuong trinh co nghiem kep: x1 = x2 = {x}')
    else:
        print('Phuong trinh vo nghiem')
```

**Bài 8:** Giải hệ phương trình tuyến tính:
$ax+by=m$
$\lfloor cx+dy=n$

```python
a, b, m = map(float, input().split())
c, d, n = map(float, input().split())
#Tính theo định thức của x,y và định thức của ma trận hệ số
D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c

if D != 0:
    x = Dx / D
    y = Dy / D
    print(f'Nghiem cua he phuong trinh la x={x} , y={y}')
else:
    if Dx == 0 and Dy == 0:
        print('He phuong trinh vo so nghiem')
    else:
        print('He phuong trinh vo nghiem')
```

**Bài 10:** Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm $A(x,y)$ bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
tam1 , tam2 = map(float, input().split())
r1=float(input())
a1 , a2 = map(float,input().split())
r=((a1-tam1)**2+(a2-tam2)**2)**0.5
if r>r1:
    print('Diem A nam ngoai duong tron')
elif r==r1:
    print('Diem A nam tren duong tron')
else:
    print("Diem A nam trong duong tron")
```
