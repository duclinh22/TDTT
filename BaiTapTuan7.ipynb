Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần. Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. mẫu hàm: def binary_search(arr, left, right, target):

```python
def binary_search(arr,left,right,target): #Tim kiem phan tu target trong list arr,tu ben trai left den ben phai right
    while True:    
        mid=(left+right)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            left=mid+1
        elif arr[mid]>target:
            right=mid-1
        if left>right:
            return -1
data=list(map(int,input().split()))
n=int(input())
print(binary_search(data,0,len(data)-1,n))
```

**W7A7**

Cho một danh sách các số nguyên nums, hãy tìm độ dài của dãy con tăng nghiêm ngặt dài nhất (không nhất thiết phải liền kề).

```python
data=list(map(int,input().split()))
def dodaimax(nums):
    if not nums :
        return 0
    dp=[1] * len(nums)
    for i in range(1,len(nums)):
        for j in range(i):
            if nums[i]>nums[j]:
                dp[i]=max(dp[i],dp[j]+1)
    return max(dp)
print(dodaimax(data))
```

**W7A8**

Bạn được cung cấp: một danh sách các khoảng thời gian interval =[[l1,r1], [l2, r2], ...] và một danh sách các truy vấn queries =[q1, q2, ...]
Với mỗi truy vấn q, hãy tìm độ dài nhỏ nhất của khoảng thời gian [li, ri] sao cho $li\le q\le ri$.
Nếu không tồn tại khoảng thời gian nào như vậy, hãy trả về -1 cho truy vấn đó.
solve in $n<O(m.n)$, hint: binary search in the queries list

```python
intervals = eval(input())
queries = eval(input())
ket_qua = []

for q in queries:
    nho_nhat = -1
    for l, r in intervals:
        if l <= q <= r:
            do_dai = r - l + 1
            if nho_nhat == -1 or do_dai < nho_nhat:
                nho_nhat = do_dai
    ket_qua.append(nho_nhat)

print(ket_qua)
```

**W7A9**

Cho một danh sách các số nums, với mỗi phần tử nums[i], hãy đếm xem có bao nhiêu phần tử bên phải nhỏ hơn nums[i].
Trả về danh sách các số đếm.
Efficient solution: using heap or binary search tree

```python
import bisect
data=list(map(int,input().split()))
ketqua=[]
danhsachtam=[]
data1=reversed(data)
for so in data1:
    index=bisect.bisect_left(danhsachtam,so)
    ketqua.append(index)
    bisect.insort(danhsachtam,so)
ketqua.reverse()
print(ketqua)
```

**W7A10**

Cho một danh sách các số nguyên nums và một số nguyên m, hãy tìm tổng mảng con lớn nhất theo modulo m.
Tức là, với bất kỳ mảng con nums[i:j] nào, hãy tính sum(nums[i:j]) % m và trả về kết quả lớn nhất có thể.
Gợi ý:
Duy trì tổng tiền tố (prefix sum) mod m và sử dụng tìm kiếm nhị phân để tìm tổng tiền tố nhỏ nhất chỉ lớn hơn tổng hiện tại một chút (để tối đa hóa hiệu số bao quanh).

```python
import bisect

nums = list(map(int, input().split()))
m = int(input())

tong_hien_tai = 0
lon_nhat = 0
danh_sach_tong = []

for so in nums:
    tong_hien_tai = (tong_hien_tai + so) % m
    
    if tong_hien_tai > lon_nhat:
        lon_nhat = tong_hien_tai
        
    vi_tri = bisect.bisect_right(danh_sach_tong, tong_hien_tai)
    
    if vi_tri < len(danh_sach_tong):
        gia_tri_moi = (tong_hien_tai - danh_sach_tong[vi_tri] + m) % m
        if gia_tri_moi > lon_nhat:
            lon_nhat = gia_tri_moi
            
    bisect.insort(danh_sach_tong, tong_hien_tai)

print(lon_nhat)
```

**W7A13**

Dãy hài hòa là một dãy thỏa mãn điều kiện giá trị lớn nhất và giá trị nhỏ nhất của dãy khác nhau đúng bằng $1$.
Viết hàm trả về độ dài của dãy hài hòa con dài nhất có thể có trong mảng giá trị nguyên $arr$ có $n$ phần tử
Ví dụ:
Input: 1 3 2 2 5 2 3 7
Output: 5 - Dãy hài hòa con dài nhất là [3,2,2,2,3]

```python
data=list(map(int,input().split()))
st=set(data)
dem=0
lst1=list(st)
lst1.sort
lst=[]
for i in lst1:
    cont=data.count(i)
    lst.append(cont)
for i in range(len(lst)-1):
    if lst[i]+lst[i+1]>dem:
        dem=lst[i]+lst[i+1]
print(dem)
```
