**Bài 2:** Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân. [cite: 3, 4, 5, 6]

```python
a=int(input('Nhập chiều dài'))
b=int(input('Nhập chiều rộng'))
dientich=a*b
pi=float(3.14)
dientichhinhtron=pi*2*b
dientichconlai=dientich - dientichhinhtron
print(dientichconlai)
```

**Bài 3:** Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem trước cấu trúc lựa chọn: if....else, ma). [cite: 7, 8]

```python
c= str(input('VIết ký tự vào đây'))
if c.isupper():
    print(c.lower()) 
elif c.islower():
    print(c.upper())  
else:
    print("Loi roi!!!")
```
