### W4A23

Cho $S(k)=1+2+3+...+k$
Tim k sao cho $S(k)$ lớn nhất nhưng nhỏ hơn n

```python
n=int(input())
k=0
while (k*(k+1))/2<n:
    k+=1
print(k)
```

### W4A24
Nhập vào A, tìm n nhỏ nhất sao cho
$1+1/2+1/3+1/4+...+1/n>A$

```python
a=int(input())
tong=1
n=1
while tong<=a:
    n+=1
    tong+=1/n
print(n)
```

### W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
n=0
max=0
min=0
while n!=-1:
    n=int(input())
    if n>max:
        max=n
    if n<min:
        min=n
print(max,min)
```

### W4A26 [Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quả A

```python
a=1
b=1
c=1
n=int(input())
while (a+b)<=n:
    c=a
    a+=b
    b=c
print(a)
```

### W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số

```python
n=input()
hoa=0
thuong=0
so=0
for i in range(len(n)):
    if n[i].isupper():
        hoa+=1
    if n[i].islower():
        thuong+=1
    if n[i].isdigit():
        so+=1
    else:
        pass
print(hoa,thuong,so)
    
```
