**W5A19**
Viết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_magic_number(n):
    n = abs(n)
    while n > 9:
        tong = 0
        for chu_so in str(n):
            tong += int(chu_so)
        n = tong
    return n == 1
```

```python
def quadratic_roots_count(a,b,c):
    if a==b==0
        return 0
    elif a==0:
        return 1
    elif b**2-4*a*c>0:
        return 2
    elif b**2-4*a*c==0:
        return 1
    else:
        return 0
```

**W5A25**
Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), (0,3)], p=(2,1)

```python
def dien_tich_tam_giac(x1, y1, x2, y2, x3, y3):
    return abs(x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) / 2

def point_in_quadrilateral(quad, p):
    x, y = p
    x1, y1 = quad[0]
    x2, y2 = quad[1]
    x3, y3 = quad[2]
    x4, y4 = quad[3]

    s_tu_giac = dien_tich_tam_giac(x1, y1, x2, y2, x3, y3) + dien_tich_tam_giac(x1, y1, x3, y3, x4, y4)
    
    s1 = dien_tich_tam_giac(x, y, x1, y1, x2, y2)
    s2 = dien_tich_tam_giac(x, y, x2, y2, x3, y3)
    s3 = dien_tich_tam_giac(x, y, x3, y3, x4, y4)
    s4 = dien_tich_tam_giac(x, y, x4, y4, x1, y1)
    
    s_p = s1 + s2 + s3 + s4

    return abs(s_tu_giac - s_p) < 0.0001
```

**W5A28**
Viết hàm tính góc ZABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math

def angle_ABC(x1, y1, x2, y2, x3, y3):
    a = math.sqrt((x1 - x2)**2 + (y1 - y2)**2)
    b = math.sqrt((x3 - x2)**2 + (y3 - y2)**2)
    c = math.sqrt((x1 - x3)**2 + (y1 - y3)**2)
    
    if a == 0 or b == 0:
        return 0
        
    cos_B = (a**2 + b**2 - c**2) / (2 * a * b)
    goc_rad = math.acos(cos_B)
    
    return math.degrees(goc_rad)
```

**W5A29**
Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].

```python
def max_min_list(a, b, c):
    lon_nhat = a
    if b > lon_nhat:
        lon_nhat = b
    if c > lon_nhat:
        lon_nhat = c
        
    nho_nhat = a
    if b < nho_nhat:
        nho_nhat = b
    if c < nho_nhat:
        nho_nhat = c
        
    return [lon_nhat, nho_nhat]
```

**W5A30**
Viết hàm tính hệ số nhị thức C(n,k).

```python
def binomial_coefficient(n, k):
    tu_so = 1
    for i in range(1, n + 1):
        tu_so *= i
        
    mau_so_1 = 1
    for i in range(1, k + 1):
        mau_so_1 *= i
        
    mau_so_2 = 1
    for i in range(1, n - k + 1):
        mau_so_2 *= i
        
    return tu_so // (mau_so_1 * mau_so_2)
```

**W5A33**
Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    gia_mo_cua = 10000
    
    if car_type == '4_cho':
        gia_km = 15000
    else:
        gia_km = 17000
        
    if km <= 1:
        tien = gia_mo_cua
    else:
        tien = gia_mo_cua + (km - 1) * gia_km
        
    tien += wait_minutes * 500
    
    if start_hour >= 22 or start_hour <= 5:
        tien += tien * 0.1
        
    if km > 30:
        tien -= tien * 0.2
        
    return tien
```
