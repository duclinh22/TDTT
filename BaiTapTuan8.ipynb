Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu mỗi phần tử x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.

```python
data=list(map(int,input().split()))
for i in data:
    if data.count(i)==i:
        pass
    elif data.count(i)>i:
        for i in range(data.count(i)-i):
            data.remove(i)
    elif data.count(i)<i:
        while i in data:
            data.remove(i)
print(data)
```

**W7A25**

Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số

```python
data=input()
k=int(input())
def timmax(data,k):
    stack=[]
    drop=len(data)-k
    for i in data:
        while stack and drop>0 and stack[-1]<i:
            stack.pop()
            drop-=1
        stack.append(i)
    return ''.join(stack[:k])
print(timmax(data,k))
```

Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ hay không

```python
data=input()
lst=[]
def check(data,lst):
    for i in data:
        if i in ['{},'[],'()']:
            lst.append(i)
        else:
            if not lst:
                return False
            match i:
                case '}':
                    op='{'
                case ']':
                    op='['
                case ')':
                    op='('
            cl=lst.pop()
            if cl != op:
                return False
    return not lst
```

Bài toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên nào $<=N$ (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua)

```python
data = input()
n = int(input())
daychuoi = ''
lenmax = 0
i = 0

while i < len(data):
    if int(data[i]) <= n:
        i += 1
        continue
    j = 1
    while i + j < len(data) and int(data[i+j]) > n:
        j += 1        
    if j > lenmax:
        lenmax = j
        daychuoi = data[i:i+j] 
        
    i += j

print(lenmax)
print(daychuoi)
```

Tìm Dây con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập vào từ bàn phím

```python
data=input()
lst=[]
for i in data:
    lst.append([int(i)])
for i in range(len(lst)):
    lentamthoi=0
    indextamthoi=0
    for j in range(i):
        if lst[j][-1]<lst[i][0]:
            if len(lst[j])>lentamthoi:
                lentamthoi=len(lst[j])
                indextamthoi=j
    if lentamthoi!=0:
        lst[i]=lst[indextamthoi]+lst[i]
lst.sort(key=len)
print(lst[-1])
```

**W7A31**

Đổi chỗ đúng 2 chữ số để được số nhỏ nhất có thể

```python
def solve():
    s = list(input().strip())
    n = len(s)
    if n < 2:
        print("".join(s))
        return
    for i in range(n):
        min_val = s[i]
        min_idx = -1
        for j in range(n - 1, i, -1):
            if i == 0 and s[j] == '0':
                continue
            if s[j] < min_val:
                min_val = s[j]
                min_idx = j
        if min_idx != -1:
            s[i], s[min_idx] = s[min_idx], s[i]
            print("".join(s))
            return

    for i in range(n - 1):
        if s[i] == s[i+1]:
            print("".join(s)) 
            return
    s[-1], s[-2] = s[-2], s[-1]
    
    if s[0] == '0':
        s[-1], s[-2] = s[-2], s[-1] 
        
    print("".join(s))

solve()
```
