```python
n=int(input())
sobuoc=0
while n!=int(str(n)[::-1]): #chuyển n sang str xong đảo ngược ký tự rồi chuyển về số
    sobuoc+=1
    n=n+int(str(n)[::-1])
print(n,sobuoc)
```

"Vừa gà vừa chó,
Bó lại cho tròn,
Ba mươi sáu con,
Một trăm chân chẫn"
Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và số lượng chó. Ngược lại, in ra "invalid".

```python
m,n=map(int,input().split())
uocchung=0
max=0
if m>n:
    max=m
else:
    max=n
for i in range(1,max):
    if m%i==0 and n%i==0:
        uocchung=i
print(uocchung)
```
