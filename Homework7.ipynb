**W7A17**

Tìm phần tử xuất hiện nhiều hơn $n/2$ lần trong mảng (nếu có) với độ phức tạp thời gian là $O(n)$, độ phức tạp bộ nhớ là $O(1)$, - Boyer-Moore. Nếu không có phần tử nào thỏa mãn thì không in gì

```python
def find_majority_element(nums):
    candidate=None
    count=0
    for num in nums:
        if count==0:
            candidate=num
            count=1
        elif num==candidate:
            count+=1
        else:
            count-=1
    check=0
    for i in nums:
        if i==candidate:
            check+=1
    if check>len(nums)//2:
        return candidate
    else:
        return None
data=list(map(int,input().split()))
print(find_majority_element(data))
```

```python
data=input()
k=int(input())
xoa=len(data)-k
stack=[]
for x in data:
    while xoa>0 and len(stack)>0 and stack[-1]<x:
        stack.pop()
        xoa-=1
    stack.append(x)
while xoa>0:
    stack.pop()
    xoa-=1
ketqua=''.join(stack)
print(ketqua)
    
```

**W7A27**

Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ hay không

```python
def kiem_tra_ngoac_hop_le(s):
    stack = []
    mapping = {")": "(", "}": "{", "]": "["}

    for char in s:
        if char in mapping:
            if not stack:
                return False
            
            top_element = stack.pop()
            if mapping[char] != top_element:
                return False
        else:
            stack.append(char)

    if not stack:
        return True
    else:
        return False

day_ngoac_1 = "()[]{}"
day_ngoac_2 = "([)]"

print(kiem_tra_ngoac_hop_le(day_ngoac_1))
print(kiem_tra_ngoac_hop_le(day_ngoac_2))
```

**W7A29**

Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau

```python
def tim_day_tang_dan(arr):
    if len(arr)==0:
        return []
    max_len=1
    len_now=1
    end_index=0
    for i in range(1,len(arr)):
        if arr[i]>arr[i-1]:
            len_now+=1
        else:
            len_now=1
        if len_now>max_len:
            max_len=len_now
            end_index=i #Luu lai xem chay den so nao roi
    start_index=end_index-max_len+1
    return arr[start_index:end_index+1]
data=list(map(int,input().split()))
print(tim_day_tang_dan(data))
```

**W7A30**

Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập vào từ bàn phím

```python
data=list(map(int,input().split()))
def timdaycontangdan(arr):
    n=len(arr)
    l=[]
    for i in range(n):
        l.append([arr[i]])
    for i in range(n):
        for j in range(i):
            if arr[i]>arr[j]:
                if len(l[j])+1>len(l[i]):
                    l[i]=l[j]+[arr[i]]
    daydainhat=[]
    for i in l:
        if len(i)>len(daydainhat):
            daydainhat=i
    return daydainhat  
print(timdaycontangdan(data))
```
