### W6A2
Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự

```python
n=list(map(int,input().split()))
lst=[]
tong=0
for i in n:
    tong+=int(i)
    lst.append(tong)
print(lst)
```

### W6A3
Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kê). In ra tuple sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)

```python
lst=list(map(int,input().split()))
k=int(input())
n=len(lst)
lst_kq=[]
if k>n:
    k=k%n
for i in range(k,n):
    lst_kq.append(lst[i])
for i in range(0,k):
    lst_kq.append(lst[i])
print(lst_kq)
```
